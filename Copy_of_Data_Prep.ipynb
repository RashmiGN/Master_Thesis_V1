```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

model = SentenceTransformer('all-MiniLM-L6-v2')

category_embeddings = model.encode(categories_df['cleaned_name'].tolist(), convert_to_tensor=True)

similarity_matrix = util.cos_sim(category_embeddings, category_embeddings)

threshold = 0.8
clusters = {}
for idx, name in enumerate(categories_df['cleaned_name']):
    for jdx, score in enumerate(similarity_matrix[idx]):
        if score > threshold and idx != jdx:
            clusters.setdefault(name, []).append(categories_df['cleaned_name'].iloc[jdx])

clusters_df = pd.DataFrame([(root, sub) for root, subs in clusters.items() for sub in subs],
                           columns=['Root', 'Subcategory'])
clusters_df.to_csv("category_clusters.csv", index=False)


print(clusters_df.head(10))
```

## **Agglomerative clustering**

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from collections import Counter


from google.colab import drive
drive.mount('/content/drive')


file_path = '/content/drive/My Drive/Colab Notebooks/category_clusters.csv'

categories_df = pd.read_csv(file_path)

categories_df = categories_df[categories_df['Subcategory'].str.strip() != '']

category_names = categories_df['Subcategory'].astype(str).tolist()


vectorizer = TfidfVectorizer(stop_words='english')
embeddings = vectorizer.fit_transform(category_names)

n_clusters = 310
clustering_model = AgglomerativeClustering(n_clusters=n_clusters, metric='cosine', linkage='average')


cluster_labels = clustering_model.fit_predict(embeddings.toarray())


categories_df['group_label'] = cluster_labels


def refine_root_node_name(group):
    terms = ' '.join(group).split()
    term_counts = Counter(terms)
    top_terms = [term for term, _ in term_counts.most_common(2)]
    return ' & '.join(top_terms)


grouped_categories = categories_df.groupby('group_label')['Subcategory'].apply(list)

root_node_mapping = {group: refine_root_node_name(subcategories) for group, subcategories in grouped_categories.items()}


categories_df['root_node'] = categories_df['group_label'].map(root_node_mapping)

grouped_categories.to_csv("grouped_categories.csv", index=False)
categories_df.to_csv("refined_clusters_310.csv", index=False)


print("Refined Clusters:")
for group, subcategories in grouped_categories.items():
    print(f"Group {group} (Root Node: {root_node_mapping[group]}): {subcategories}")
    if group > 10:
        break


def generalize_root_node_name(root_node):

    generalization_map = {
        "starters & starter": "Starters & Systems",
        "machines & motors": "Industrial Machines",
        "celebrities & parker": "Celebrities & Entertainment",
        "other & lots & bags": "Retail & Mixed Goods",
        "food & nuts": "Food & Beverages",

    }

    return generalization_map.get(root_node, root_node)


categories_df['generalized_root_node'] = categories_df['root_node'].apply(generalize_root_node_name)


categories_df.to_csv("generalized_clusters.csv", index=False)
print("Dataset with generalized root node names saved to 'generalized_clusters.csv'.")

grouped_categories_generalized = categories_df.groupby(['group_label', 'generalized_root_node'])['Subcategory'].apply(list)


for (group, root_node), group_items in grouped_categories_generalized.items():
    print(f"Group {group} (Generalized Root Node: {root_node}): {group_items}")
    print("-" * 100)
    if group > 10:
        break
```
