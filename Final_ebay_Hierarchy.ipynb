```python
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset

torch.random.manual_seed(0)


dataset_name = input("Enter the name of the dataset from Huggingfaces(e.g., RashmiNaik/ebay_products_keywords_dataset): ")


model_name = "microsoft/Phi-3.5-mini-instruct"
print(f"Loading model '{model_name}'...")
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="cuda", torch_dtype="auto", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)


generator = pipeline("text-generation", model=model, tokenizer=tokenizer)


print(f"Loading dataset '{dataset_name}'...")
try:
    dataset = load_dataset(dataset_name)
    print(f"Dataset '{dataset_name}' loaded successfully.")
except Exception as e:
    print(f"Failed to load dataset '{dataset_name}': {e}")
```

```python
import time

# LEVEL 1
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from collections import Counter
import re

df = pd.DataFrame(dataset['train'])


df['Associated Keywords'] = df['Associated Keywords'].apply(lambda x: x if isinstance(x, list) else x.split(', '))
df['Keyword Sets'] = df['Associated Keywords'].apply(frozenset)
unique_df = df.drop_duplicates(subset=['Keyword Sets'])
unique_df['Keywords List'] = unique_df['Keyword Sets'].apply(list)

all_keywords = [keyword for keywords_set in unique_df['Keywords List'] for keyword in keywords_set]
keyword_counts = Counter(all_keywords)
duplicates = {keyword: count for keyword, count in keyword_counts.items() if count > 1}

def remove_duplicates(keyword_list, duplicates):
    return [keyword for keyword in keyword_list if keyword not in duplicates]

unique_df['Keywords List'] = unique_df['Keywords List'].apply(remove_duplicates, duplicates=duplicates)


vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(unique_df['Keywords List'].apply(' '.join))


kmeans = KMeans(n_clusters=338, random_state=42)
unique_df['Cluster ID'] = kmeans.fit_predict(X)


def extract_top_keywords(keyword_list, top_n=15):
    keyword_counts = Counter(keyword_list)
    top_keywords = keyword_counts.most_common(top_n)
    return ', '.join(keyword for keyword, _ in top_keywords)

unique_df['Top 15 Keywords'] = unique_df['Keywords List'].apply(lambda x: extract_top_keywords(x, top_n=15))


data_for_new_df = []
conversation_prefix = "<|system|>You are a helpful assistant.<|end|>"

for index, row in unique_df.iterrows():
    keywords_str = row['Top 15 Keywords']
    user_input = f"<|user|>Generate a concise 2 words relevant name for these products keywords set: {keywords_str}<|end|>"
    prompt = f"{conversation_prefix} {user_input} <|assistant|>"

    response = generator(prompt, max_length=1024, num_return_sequences=1, temperature=0.5)
    generated_text = response[0]['generated_text']
    category_name = generated_text.split('<|end|>')[-1].strip()
    category_name = category_name.replace('<|assistant|>', '').strip()

    data_for_new_df.append({
        'Cluster ID': row['Cluster ID'],
        'Product Description': row['Product Description'],
        'Keywords': row['Top 15 Keywords'],
        'Generated Name': category_name
    })

new_df = pd.DataFrame(data_for_new_df)


def clean_generated_name(name):

    removal_phrases = [
        "This category name encapsulates",
        "This name encapsulates the essence",
        "This two-word name encapsulate",
        "This name encapsulates the eco-friendly"
    ]


    for phrase in removal_phrases:
        if phrase in name:
            name = name.split(phrase)[0].strip()


    name = re.split(r'\n\d+\..*', name)[0]


    name = re.split(r'[,\n;]', name)[0].strip()


    name = re.sub(r'^\d+\.\s*', '', name)


    name = name.strip('\"\' ')


    if '.' in name and len(name.split('.')[0].split()) < 10:
        name = name.split('.')[0].strip()

    return name

new_df['Category I'] = new_df['Generated Name'].apply(clean_generated_name)
First_level_df = new_df[['Cluster ID', 'Product Description','Keywords', 'Category I']]


time.sleep(20)

# LEVEL 2
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(First_level_df['Category I'])

kmeans = KMeans(n_clusters=120, random_state=42)
kmeans.fit(X)
First_level_df['New Cluster ID'] = kmeans.labels_

misc_categories = ['Eyewear Trends', 'Casual & Formal', 'Decorative Trims', 'Dog Breeds & Portraits',
                   'Casual & Outerwear', 'Gourmet & Organic', 'Anime Merchandise', 'Casual/Formal Shoes',
                   'Dress and Casual Shoes', 'Robotic Assistants', 'Dancewear & Barres', 'Chisel Types',
                   'Studio Monitors', 'Navigation Robots', 'Pool Maintenance & Filtration', 'Collectible Figurines', 'Holiday Decorations']

misc_cluster_id = First_level_df['New Cluster ID'].max() + 1
First_level_df.loc[First_level_df['Category I'].isin(misc_categories), 'New Cluster ID'] = misc_cluster_id

cluster_summary = First_level_df.groupby('New Cluster ID')['Category I'].apply(lambda names: ', '.join(set(names))).reset_index()
cluster_summary.columns = ['New Cluster ID', 'Clustered Categories']
cluster_summary.loc[cluster_summary['New Cluster ID'] == misc_cluster_id, 'Clustered Categories'] = 'Miscellaneous'

second_level_df = First_level_df.merge(cluster_summary, on='New Cluster ID', how='left')
second_level_df = second_level_df[['Cluster ID', 'Keywords', 'Category I', 'Product Description', 'Clustered Categories']]

unique_clusters = second_level_df['Clustered Categories'].unique()
data_for_cluster_names = []

for category in unique_clusters:
    user_input = f"<|user|>Generate a concise 2 words relevant name for these products keywords set: {category}<|end|>"
    prompt = f"{conversation_prefix} {user_input} <|assistant|>"

    response = generator(prompt, max_length=1024, num_return_sequences=1, temperature=0.5, do_sample=True)
    generated_text = response[0]['generated_text']
    generated_name = generated_text.split('<|end|>')[-1].strip().replace('<|assistant|>', '').strip()

    data_for_cluster_names.append({
        'Clustered Categories': category,
        'Generated Category II Names': generated_name
    })

generated_names_df = pd.DataFrame(data_for_cluster_names)
new_second_level_df = pd.merge(second_level_df, generated_names_df, on='Clustered Categories', how='left')

new_second_level_df['Category II'] = new_second_level_df['Generated Category II Names'].apply(clean_generated_name)
new_second_level_df['Category II'] = new_second_level_df['Category II'].apply(lambda x: x if len(x.split()) == 2 else 'Miscellaneous')
new_second_level_df = new_second_level_df[['Cluster ID', 'Keywords', 'Product Description', 'Category I', 'Category II']]


time.sleep(20)

# LEVEL 3
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(new_second_level_df['Category II'])

kmeans = KMeans(n_clusters=40, random_state=42)
kmeans.fit(X)
new_second_level_df['New Cluster ID'] = kmeans.labels_

misc_cluster_id = new_second_level_df['New Cluster ID'].max() + 1
new_second_level_df.loc[new_second_level_df['Category II'].isin(misc_categories), 'New Cluster ID'] = misc_cluster_id

cluster_summary = new_second_level_df.groupby('New Cluster ID')['Category II'].apply(lambda names: ', '.join(set(names))).reset_index()
cluster_summary.columns = ['New Cluster ID', 'Clustered Categories III']

final_df = new_second_level_df.merge(cluster_summary, on='New Cluster ID', how='left')

data_for_cluster_names = []
for category in final_df['Clustered Categories III'].unique():
    user_input = f"<|user|>Generate a concise 2 words relevant name for these products keywords set: {category}<|end|>"
    prompt = f"{conversation_prefix} {user_input} <|assistant|>"

    response = generator(prompt, max_length=1024, num_return_sequences=1, temperature=0.5, do_sample=True)
    generated_text = response[0]['generated_text']
    generated_name = generated_text.split('<|end|>')[-1].strip().replace('<|assistant|>', '').strip()

    data_for_cluster_names.append({
        'Clustered Categories III': category,
        'Generated Category III Names': generated_name
    })

generated_names_df = pd.DataFrame(data_for_cluster_names)
final_df = pd.merge(final_df, generated_names_df, on='Clustered Categories III', how='left')

final_df['Category III'] = final_df['Generated Category III Names'].apply(clean_generated_name)
final_df['Category III'] = final_df['Category III'].apply(lambda x: x if len(x.split()) == 2 else 'Miscellaneous')
final_df = final_df[['Cluster ID', 'Keywords', 'Product Description', 'Category I', 'Category II', 'Category III']]
```

```python
import pandas as pd
import plotly.graph_objects as go


hierarchy_data = final_df.groupby(['Category III', 'Category II', 'Category I']).size().reset_index(name='Count')

table_rows = []

for category_iii in hierarchy_data['Category III'].unique():

    table_rows.append([category_iii, '', '', ''])


    category_ii_data = hierarchy_data[hierarchy_data['Category III'] == category_iii]

    for category_ii in category_ii_data['Category II'].unique():

        table_rows.append(['', category_ii, '', ''])


        category_i_data = category_ii_data[category_ii_data['Category II'] == category_ii]

        for _, row in category_i_data.iterrows():

            table_rows.append(['', '', row['Category I'], row['Count']])


table_df = pd.DataFrame(table_rows, columns=['Category III', 'Category II', 'Category I', 'Count'])


fig = go.Figure(data=[go.Table(
    header=dict(
        values=['<b>Category III</b>', '<b>Category II</b>', '<b>Category I</b>', '<b>Count</b>'],
        fill_color='lightgrey',
        align='left'
    ),
    cells=dict(
        values=[table_df[col] for col in table_df.columns],
        fill_color='white',
        align='left'
    )
)])


fig.update_layout(
    title='Hierarchical Category Table',
    height=600,
    margin=dict(t=50, l=25, r=25, b=25)
)


fig.show()
```
