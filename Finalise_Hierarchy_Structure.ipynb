```python
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer

torch.random.manual_seed(0)

model_name = "microsoft/Phi-3.5-mini-instruct"
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="cuda", torch_dtype="auto", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from collections import Counter


df['Associated Keywords'] = df['Associated Keywords'].apply(lambda x: x if isinstance(x, list) else x.split(', '))

df['Keyword Sets'] = df['Associated Keywords'].apply(frozenset)
unique_df = df.drop_duplicates(subset=['Keyword Sets'])


unique_df['Keywords List'] = unique_df['Keyword Sets'].apply(list)

all_keywords = [keyword for keywords_set in unique_df['Keywords List'] for keyword in keywords_set]
keyword_counts = Counter(all_keywords)

duplicates = {keyword: count for keyword, count in keyword_counts.items() if count > 1}

def remove_duplicates(keyword_list, duplicates):
    return [keyword for keyword in keyword_list if keyword not in duplicates]

unique_df['Keywords List'] = unique_df['Keywords List'].apply(remove_duplicates, duplicates=duplicates)

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(unique_df['Keywords List'].apply(' '.join))

kmeans = KMeans(n_clusters=338, random_state=42)
unique_df['Cluster ID'] = kmeans.fit_predict(X)

def extract_top_keywords(keyword_list, top_n=15):
    keyword_counts = Counter(keyword_list)
    top_keywords = keyword_counts.most_common(top_n)
    return ', '.join(keyword for keyword, _ in top_keywords)

unique_df['Top 15 Keywords'] = unique_df['Keywords List'].apply(lambda x: extract_top_keywords(x, top_n=15))
```

```python
import pandas as pd

data_for_new_df = []

conversation_prefix = "<|system|>You are a helpful assistant.<|end|>"

for index, row in unique_df.iterrows():
    keywords_str = row['Top 15 Keywords']
    user_input = f"<|user|>Generate a concise 2 words relevant name for these products keywords set: {keywords_str}<|end|>"
    prompt = f"{conversation_prefix} {user_input} <|assistant|>"

    response = generator(prompt, max_length=1024, num_return_sequences=1, temperature=0.5)
    generated_text = response[0]['generated_text']
    category_name = generated_text.split('<|end|>')[-1].strip()

    category_name = category_name.replace('<|assistant|>', '').strip()

    # Collect each row of new data
    data_for_new_df.append({
        'Cluster ID': row['Cluster ID'],
        'Product Description': row['Product Description'],
        'Keywords': row['Top 15 Keywords'],
        'Generated Name': category_name
    })


new_df = pd.DataFrame(data_for_new_df)


def clean_generated_name(name):

    removal_phrases = [
        "This category name encapsulates",
        "This name encapsulates the essence",
        "This two-word name encapsulate",
        "This name encapsulates the eco-friendly"
    ]

    # Remove phrases from the name
    for phrase in removal_phrases:
        if phrase in name:
            name = name.split(phrase)[0].strip()
    entries = name.split(',')
    name = next((entry.strip() for entry in entries if entry.strip().startswith("1.")), name)
    name = name.split('.', 1)[-1].strip() if '.' in name else name
    name = name.strip('\"')

    return name

new_df['Category I'] = new_df['Generated Name'].apply(clean_generated_name)

First_level_df = new_df[['Cluster ID', 'Top 15 Keywords', 'Generated Name', 'Category I']]

```
