````python
import torch
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    pipeline,
)

MODEL_NAME = "RashmiNaik/Llama-3-8B-Instruct-taxonomy-constructor"
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True, bnb_4bit_quant_type="nf4", bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME, quantization_config=quantization_config, device_map="auto"
)
````

````python
def format_example(row: dict):
    prompt = dedent(
        f"""
    {row["question"]}

    Information:

    ```
    {row["context"]}
    ```
    """
    )
    messages = [
        {
            "role": "system",
            "content": "Use only the information to answer the question",
        },
        {"role": "user", "content": prompt},
        {"role": "assistant", "content": row["answer"]},
    ]
    return tokenizer.apply_chat_template(messages, tokenize=False)
````

````python
pipe = pipeline(
    task="text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=250,
    return_full_text=False,
)
````

````python
import pandas as pd
import ast
from datasets import load_dataset
from textwrap import dedent
from collections import defaultdict
from IPython.core.display import display, HTML

dataset = load_dataset("RashmiNaik/ebay_new_inputs")["train"]

def create_test_prompt(question, context):
    prompt = dedent(
        f"""
        {question}

        Information:

        ```
        {context}
        ```
        """
    )
    messages = [
        {
            "role": "system",
            "content": "Use only the information to answer the question",
        },
        {"role": "user", "content": prompt},
    ]
    return tokenizer.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )


def process_dataset(dataset):
    results = []

    for idx, question in enumerate(dataset["inputs"]):
        context = "Based on the Product Description given, categorize the item and generate appropriate taxonomy."


        prompt = create_test_prompt(question, context)


        outputs = pipe(prompt)


        generated_text = outputs[0]['generated_text']

        try:
            response_dict = ast.literal_eval(generated_text)
        except Exception as e:
            print(f"Error parsing response at index {idx}: {e}")
            response_dict = {}


        row_data = {
            "inputs": question,
            "Category I": response_dict.get("Primary Category", "Miscellaneous"),
            "Category II": response_dict.get("Parent Category", "Miscellaneous"),
            "Category III": response_dict.get("General Category", "Miscellaneous"),
        }
        results.append(row_data)


    print("Dataset processing complete.")
    return pd.DataFrame(results)


df = process_dataset(dataset)

category_hierarchy = df.groupby(["inputs", "Category I", "Category II"])["Category III"].unique().reset_index()

taxonomy_tree = defaultdict(lambda: defaultdict(list))

for _, row in category_hierarchy.iterrows():
    product_description = row["inputs"]
    category_i = row["Category I"]
    category_ii = row["Category II"]
    category_iii = row["Category III"][0]

    taxonomy_tree[category_iii][category_ii].append((category_i, product_description))

taxonomy_data = []
for category_iii, subcategories in taxonomy_tree.items():
    for category_ii, category_is in subcategories.items():
        for category_i, product_description in category_is:
            taxonomy_data.append([category_iii, category_ii, category_i, product_description])

taxonomy_df = pd.DataFrame(taxonomy_data, columns=["Category III", "Category II", "Category I", "Product Description"])

final_output_file = "/content/drive/My Drive/Colab Notebooks/finetuned_model_outputs.csv"
taxonomy_df.to_csv(final_output_file, index=False)


html = """
<style>
table { width: 100%; border-collapse: collapse; }
th, td { border: 1px solid #ddd; padding: 8px; text-align: left; }
th { background-color: #ccc; font-weight: bold; }
</style>
<table>
<tr><th>Category III</th><th>Category II</th><th>Category I</th><th>Product Description</th></tr>
"""

for category_iii, subcategories in taxonomy_tree.items():
    for category_ii, category_is in subcategories.items():
        for category_i, product_description in category_is:
            html += f"<tr><td>{category_iii}</td><td>{category_ii}</td><td>{category_i}</td><td>{product_description}</td></tr>"

html += "</table>"


display(HTML(html))
````

````python
import pandas as pd
from collections import defaultdict
from anytree import Node, RenderTree

file_path = "/content/drive/My Drive/Colab Notebooks/finetuned_model_outputs.csv"
df = pd.read_csv(file_path)

category_hierarchy = df.groupby(["Category I", "Category II"])["Category III"].unique().reset_index()

taxonomy_tree = defaultdict(lambda: defaultdict(set))

for _, row in category_hierarchy.iterrows():
    category_i = row["Category I"]
    category_ii = row["Category II"]
    category_iii_list = row["Category III"]

    for category_iii in category_iii_list:
        taxonomy_tree[category_iii][category_ii].add(category_i)

root = Node("ebay_Taxonomy")


category_iii_nodes = {}
category_ii_nodes = {}

for category_iii, category_ii_dict in taxonomy_tree.items():
    category_iii_node = Node(category_iii, parent=root)
    category_iii_nodes[category_iii] = category_iii_node

    for category_ii, category_i_set in category_ii_dict.items():
        category_ii_node = Node(category_ii, parent=category_iii_node)
        category_ii_nodes[category_ii] = category_ii_node

        for category_i in category_i_set:
            Node(category_i, parent=category_ii_node)

for pre, fill, node in RenderTree(root):
    print("%s%s" % (pre, node.name))
````
